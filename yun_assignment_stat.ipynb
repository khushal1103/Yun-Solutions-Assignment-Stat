```python
import requests
import pandas as pd

# Define the API endpoint URL
api_url = "https://live.markethound.in/api/history/decay"

# Parameters for the API request
params = {
    "name": "FINNIFTY",  # Index name
    "expiry": "2024-03-05T00:00:00.000Z",  # Expiry date
    "dte": "0"  # Days to expiry
}

# Make the API request
response = requests.get(api_url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse JSON response
    # Convert JSON data to DataFrame
    df = pd.DataFrame(response.json()['result'])
    # Inspect DataFrame
    print(df.head())
else:
    print("Failed to fetch data from API.")
```

```python
df = df.apply(pd.to_numeric, errors='ignore')

# Grouping and aggregating data by 'date'
df_bydate = df.groupby('date').agg({
    'high': 'max',
    'low': 'min',
    'open': lambda x: list(x)[0],
    'close': lambda x: list(x)[-1],
    'intradayMovement': lambda x: list(x)[0],
    'intradayTotal': lambda x: list(x)[0],
    'index_close': lambda x: list(x)[-1],
    'index_high': max,
    'index_low': min,
    'index_open': lambda x: list(x)[0],
    'overnightGap': lambda x: list(x)[0],
    'overnightExpected': lambda x: list(x)[0]
})

# Displaying the aggregated DataFrame
print(df_bydate)
```
